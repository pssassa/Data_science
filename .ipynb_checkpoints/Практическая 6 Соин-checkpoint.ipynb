```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder, MinMaxScaler
from category_encoders import CountEncoder
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
```

1. Загрузить файл 'telecom_churn.csv'. 

```python
df = pd.read_csv('telecom_churn.csv')
df['Area code'] = df['Area code'].astype(str)
numeric = df.select_dtypes(include='number').columns
df
```

```python
numeric
```

2. Проверить данные на наличие пропусков и выбросов. Обработать пропуски
и выбросы.

```python
# Проверка данных на пропуски
df.isna().sum()
```

```python
# Функция для обработки выбросов
def outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    return df

# Обработка выбросов для всех числовых столбцов
for col in numeric:
    df = outliers(df, col)

# Сброс индексов
df.reset_index(inplace=True, drop=True)
df
```

3. Выполнить кодирование категориальных переменных. 

```python
# Отбор категориальных переменных
categor_df = df.select_dtypes(exclude=[np.number])
categor_df.head()
```

```python
# OneHotEncoder для некоторых категориальных переменных
encoder = OneHotEncoder(drop='first', sparse_output=False)
encoded_cols = encoder.fit_transform(df[['Area code', 'International plan', 'Voice mail plan']])
encoded_df = pd.DataFrame(encoded_cols, columns=encoder.get_feature_names_out(['Area code', 'International plan', 'Voice mail plan']))
encoded_df
```

```python
# CountEncoder для переменной 'State'
c_encoder = CountEncoder(normalize=True)
state_encode = c_encoder.fit_transform(categor_df['State'])
state_encode.reset_index(inplace=True)
state_encode
```

```python
# Объединение закодированных категориальных переменных
categor_df = pd.concat([encoded_df, state_encode], axis=1)
categor_df
```

```python
# Целевая переменная
y = df['Churn']

# Объединение всех данных
concated_df = pd.concat([df, categor_df], axis=1)
concated_df.drop(['Area code', 'International plan', 'Voice mail plan', 'State', 'Churn'], axis=1, inplace=True)
concated_df
```

```python
concated_df = concated_df.drop('index', axis=1)
columns = concated_df.columns
```

4. Выполнить масштабирование признаков. 

```python
# Масштабирование признаков
m_scaler = MinMaxScaler()
rez_df = pd.DataFrame(m_scaler.fit_transform(concated_df))
rez_df.columns = columns
rez_df.head()
```

5. Построить модели k-ближайших соседей, логистической регрессии, SVM 


a. Обучить модель по кросс-валидации и выведите на экран значения метрик. 

```python
# Разделение данных на тренировочную и тестовую выборки
# Разделение данных на тренировочную и тестовую выборки
x = rez_df
y = df['Churn']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=4)

# Обучение и оценка моделей

# KNN
knn_model = KNeighborsClassifier()
knn_model.fit(x_train, y_train)
y_pred = knn_model.predict(x_test)
knn_accuracy_score = accuracy_score(y_test, y_pred)
knn_classification_report = classification_report(y_test, y_pred)
print("KNN Accuracy:", knn_accuracy_score)
print("KNN Classification Report:\n", knn_classification_report)

# Logistic Regression
logreg_model = LogisticRegression(max_iter=1000)
logreg_model.fit(x_train, y_train)
y_pred = logreg_model.predict(x_test)
log_accuracy_score = accuracy_score(y_test, y_pred)
log_classification_report = classification_report(y_test, y_pred)
print("Logistic Regression Accuracy:", log_accuracy_score)
print("Logistic Regression Classification Report:\n", log_classification_report)

# SVM
svm_model = SVC()
svm_model.fit(x_train, y_train)
y_pred = svm_model.predict(x_test)
svm_accuracy_score = accuracy_score(y_test, y_pred)
svm_classification_report = classification_report(y_test, y_pred)
print("SVM Accuracy:", svm_accuracy_score)
print("SVM Classification Report:\n", svm_classification_report)
```

b. Подобрать значение параметров модели по кросс-валидации, используя GridSearchCV. 


```python
# Параметры для подбора
knn_param_grid = {'n_neighbors': [3, 5, 7], 'weights': ['uniform', 'distance']}

# GridSearchCV для KNN
knn_grid_search = GridSearchCV(KNeighborsClassifier(), knn_param_grid, cv=5, scoring='accuracy')
knn_grid_search.fit(x_train, y_train)

# Лучшая модель KNN
best_knn_model = knn_grid_search.best_estimator_
y_pred_best_knn = best_knn_model.predict(x_test)
best_knn_accuracy = accuracy_score(y_test, y_pred_best_knn)
best_knn_class_report = classification_report(y_test, y_pred_best_knn)
print("Best KNN model parameters:", knn_grid_search.best_params_)
print("Best KNN Accuracy:", best_knn_accuracy)
print("Best KNN Classification Report:\n", best_knn_class_report)

# Параметры для подбора
logreg_param_grid = {'C': [0.01, 0.1, 1, 10, 100], 'solver': ['liblinear', 'lbfgs']}

# GridSearchCV для логистической регрессии
logreg_grid_search = GridSearchCV(LogisticRegression(max_iter=1000), logreg_param_grid, cv=5, scoring='accuracy')
logreg_grid_search.fit(x_train, y_train)

# Лучшая модель логистической регрессии
best_logreg_model = logreg_grid_search.best_estimator_
print("Best Logistic Regression model parameters:", logreg_grid_search.best_params_)

# Метрики для лучшей модели логистической регрессии
y_pred_best_logreg = best_logreg_model.predict(x_test)
best_logreg_accuracy = accuracy_score(y_test, y_pred_best_logreg)
best_logreg_class_report = classification_report(y_test, y_pred_best_logreg)
print("Best LOGREG Accuracy:", best_logreg_accuracy)
print("Best LOGREG Classification Report:\n", best_logreg_class_report)

# Параметры для подбора
svm_param_grid = {'C': [0.1, 1, 10, 100], 'gamma': [1, 0.1, 0.01, 0.001], 'kernel': ['linear', 'rbf']}

# GridSearchCV для SVM
svm_grid_search = GridSearchCV(SVC(), svm_param_grid, cv=5, scoring='accuracy')
svm_grid_search.fit(x_train, y_train)

# Лучшая модель SVM
best_svm_model = svm_grid_search.best_estimator_
print("Best SVM model parameters:", svm_grid_search.best_params_)

# Метрики для лучшей модели SVM
y_pred_best_svm = best_svm_model.predict(x_test)
best_svm_accuracy = accuracy_score(y_test, y_pred_best_svm)
best_svm_class_report = classification_report(y_test, y_pred_best_svm)
print("Best SVM Accuracy:", best_svm_accuracy)
print("Best SVM Classification Report:\n", best_svm_class_report)
```

```python
knn_report_lines = best_knn_class_report.split('\n')
logreg_report_lines = best_logreg_class_report.split('\n')
svm_report_lines = best_svm_class_report.split('\n')

knn_precision = knn_report_lines[7].split()[2]
knn_recall = knn_report_lines[7].split()[3]
knn_f1_score = knn_report_lines[7].split()[4]

logreg_precision = logreg_report_lines[7].split()[2]
logreg_recall = logreg_report_lines[7].split()[3]
logreg_f1_score = logreg_report_lines[7].split()[4]

svm_precision = svm_report_lines[7].split()[2]
svm_recall = svm_report_lines[7].split()[3]
svm_f1_score = svm_report_lines[7].split()[4]
```

```python
# Формирование словаря с метриками
metrics = {
    'Model': ['Best KNN', 'Best Logistic Regression', 'Best SVM'],
    'Accuracy': [best_knn_accuracy, best_logreg_accuracy, best_svm_accuracy],
    'Precision': [knn_precision, logreg_precision, svm_precision],
    'Recall': [knn_recall, logreg_recall, svm_recall],
    'F1-Score': [knn_f1_score, logreg_f1_score, svm_f1_score]
}

# Создание DataFrame
metrics_df = pd.DataFrame(metrics)
print(metrics_df)
```
